## Data Preprocessing

### Feature Transoformation

```python
# Transofrm MSSubClass into object 
all_data['MSSubClass'] = all_data['MSSubClass'].astype('object', copy=False)
#Verify the transformation
print all_data['MSSubClass'].dtypes
```

### Skewness of numeric features

```python
from scipy.stats import skew

###Log transform the target train
train_data['SalePrice'] = np.log1p(train_data['SalePrice'])


### Log transform numeric features
#Define the numeric feature indexes
num_idx = all_data.dtypes[all_data.dtypes != 'object'].index 
print num_idx

#determine the skewness score for each numeric features, disregarding Nan values
skewed_feats = train_data[num_idx].apply(lambda x: skew(x.dropna())) 
#select by the value of skewness if a feature which feature going to be transformed
skewed_feats = skewed_feats[skewed_feats > 0.65]
#save the name (index) of selected features
skewed_feats = skewed_feats.index

#aply log1p transformation into selected features
all_data[skewed_feats] = np.log1p(all_data[skewed_feats])
```

### Feature Selection

```python
def feature_drop_fill(df, n=75):
    """
    function to select which features will be removed, filled or is perfect filled.
    
    input
    df: dataframe of dataset; n: inter number that represents the theshold between remove and fill
    
    output
    del_col: list of features that need to be removed; fil_col: list of features that need to be filled
    """
    delete_col = []
    filling_col = []
    for col in list(df.columns.values):
        #print col, type(col)
        total_missing = df[col].isnull().sum()
        if total_missing >= n:
            print str(col)+': Missing Del count equal ',total_missing
            delete_col.append(str(col))
            #print delete_col
        elif total_missing > 0:
            #print "The feature " + str(col) + " have few missing values."
            print str(col)+': Missing Filling count equal ',total_missing
            filling_col.append(str(col))   # = filling_col.
            #print filling_col
        else:
            print str(col) + " IS PERFECT!!"
            
    return delete_col, filling_col
```

```python
#remore features with many missing values
del_col, fil_col = feature_drop_fill(all_data, n=20)

print '\n'
print "\n"
print "Number of featues deleted: ", len(del_col)
print del_col
print "\n"
print "Number of featues filled: ", len(fil_col)
print fil_col
#cat_train_data.drop(del_col, axis=1, inplace=True)
#display(cat_train_data.head(3))
```

### Remove Outlier from training set

```python
# For each feature find the data points with extreme high or low values
import math
def outliers(df, feature):
    
    #Calculate Q1 (25th percentile of the data) for the given feature
    Q1 = np.nanpercentile(df[feature], 25)
    
    # Calculate Q3 (75th percentile of the data) for the given feature
    Q3 = np.nanpercentile(df[feature], 75)
    #Use the interquartile range to calculate an outlier step (1.5 times the interquartile range)
    step = 1.5*(Q3 - Q1)
    
    if math.isnan(Q1) == True or math.isnan(Q3) == True:
        print '-'*120
        print "\n"
        print "###### Data points considered outliers for the feature '{}' ######".format(feature)
        print "\n"
        print "Q1 = {} and Q3 = {}".format(Q1, Q3)
        print "\n"
        print "\n"
        print '-'*120
        print "THERE ISN'T OUTLIERS."
        return tuple()
    
    #Outliners Indices
    out_values = df[~((df[feature] >= Q1 - step) & (df[feature] <= Q3 + step))]
    indices = out_values.index
    
    # Display the outliers
    print '-'*120
    print "\n"
    print "###### Data points considered outliers for the feature '{}' ######".format(feature)
    print "\n"
    print "Q1 = {} and Q3 = {}".format(Q1, Q3)
    print "\n"
    #display(df[~((df[feature] >= Q1 - step) & (df[feature] <= Q3 + step))])
    print "\n"
    print '-'*120
    return list(indices)
```
